```python
import pandas as pd

# Starting from your long df:
# Moca_long with columns:
# ['uid','age','timepoint','id_participant','evaluation',
#  'date_evaluation','moca_scolarite_12_annees_ou_moins_1',
#  'moca_score_total_30','moca_score_total_plus_scolarite_30']

# 1) Map timepoint codes to your labels
tp_map = {
    't00': 'Initiale',
    't02': 'suivi-2ans',
    't04': 'suivi-4ans',
    't06': 'suivi-6ans',
    't08': 'suivi-8ans',
    't10': 'suivi-10ans',  # in case you have t10
}

moca_tmp = Moca_long.copy()
moca_tmp['tp_label'] = moca_tmp['timepoint'].map(tp_map)

# 2) Pivot: age + MOCA scores by id_participant × tp_label
moca_wide = (
    moca_tmp
    .pivot_table(
        index='id_participant',
        columns='tp_label',
        values=['age',
                'moca_score_total_30',
                'moca_score_total_plus_scolarite_30']
    )
)

# 3) Flatten the MultiIndex columns: (variable, timepoint) -> 'variable_timepoint'
moca_wide.columns = [f'{var}_{tp}' for var, tp in moca_wide.columns]

# 4) Put id_participant back as a column
moca_wide = moca_wide.reset_index()
```

```python
import pandas as pd
import numpy as np

# Make sure keys are same dtype
Moca['PSCID'] = Moca['PSCID'].astype(int)
moca_wide_filtered['id_participant'] = moca_wide_filtered['id_participant'].astype(int)

merged = (
    Moca
    .rename(columns={'PSCID': 'id_participant'})
    .merge(
        moca_wide_filtered,
        on='id_participant',
        how='outer',
        indicator=True,
        suffixes=('_moca', '_wide')
    )
)

# Who didn't join?
only_in_moca = merged.loc[merged['_merge'] == 'left_only', 'id_participant'].unique()
only_in_wide = merged.loc[merged['_merge'] == 'right_only', 'id_participant'].unique()

print("Participants only in original Moca:", only_in_moca)
print("Participants only in moca_wide_filtered:", only_in_wide)
```

```python
import pandas as pd
import numpy as np

# ---- 1. Define column pairs to compare ----
pairs = {
    # Ages
    'age_Initiale':      ('Initiale',        'age_Initiale'),
    'age_suivi_2':       ('suivi-2ans',      'age_suivi-2ans'),
    'age_suivi_4':       ('suivi-4ans',      'age_suivi-4ans'),
    'age_suivi_6':       ('suivi-6ans',      'age_suivi-6ans'),
    'age_suivi_8':       ('suivi-8ans',      'age_suivi-8ans'),
    'age_suivi_10':      ('suivi-10ans',     'age_suivi-10ans'),

    # MoCA total
    'moca_total_Initiale': ('Initiale.1',    'moca_score_total_30_Initiale'),
    'moca_total_s2':       ('suivi-2ans.1',  'moca_score_total_30_suivi-2ans'),
    'moca_total_s4':       ('suivi-4ans.1',  'moca_score_total_30_suivi-4ans'),
    'moca_total_s6':       ('suivi-6ans.1',  'moca_score_total_30_suivi-6ans'),
    'moca_total_s8':       ('suivi-8ans.1',  'moca_score_total_30_suivi-8ans'),
    'moca_total_s10':      ('suivi-10ans.1', 'moca_score_total_30_suivi-10ans'),

    # MoCA + schooling
    'moca_plus_Initiale':  ('Initiale.2',    'moca_score_total_plus_scolarite_30_Initiale'),
    'moca_plus_s2':        ('suivi-2ans.2',  'moca_score_total_plus_scolarite_30_suivi-2ans'),
    'moca_plus_s4':        ('suivi-4ans.2',  'moca_score_total_plus_scolarite_30_suivi-4ans'),
    'moca_plus_s6':        ('suivi-6ans.2',  'moca_score_total_plus_scolarite_30_suivi-6ans'),
    'moca_plus_s8':        ('suivi-8ans.2',  'moca_score_total_plus_scolarite_30_suivi-8ans'),
    'moca_plus_s10':       ('suivi-10ans.2', 'moca_score_total_plus_scolarite_30_suivi-10ans'),
}

```

```python

# ---- 2. Compare values and detect ONLY mismatches ----
both_mask = merged['_merge'] == 'both'

mismatch_summaries = {}
mismatch_rows = {}   # optional: to inspect actual rows later

for name, (col_moca, col_wide) in pairs.items():
    if col_moca not in merged.columns or col_wide not in merged.columns:
        print(f"Skipping {name}: {col_moca} or {col_wide} not in merged")
        continue

    left  = pd.to_numeric(merged[col_moca], errors='coerce')
    right = pd.to_numeric(merged[col_wide], errors='coerce')

    # Only compare rows where both are non-NaN
    mask_diff = both_mask & left.notna() & right.notna() & (left != right)

    n_diff = int(mask_diff.sum())
    mismatch_summaries[name] = n_diff

    # Store detailed rows for this variable (optional)
    mismatch_rows[name] = merged.loc[mask_diff, ['id_participant', col_moca, col_wide]]

# ---- 3. Print a clean summary ----
print("Number of mismatches per variable (ignoring missing values):")
for name, n in mismatch_summaries.items():
    print(f"{name}: {n}")

# If you want a quick check: is everything identical?
any_mismatch = any(n > 0 for n in mismatch_summaries.values())
print("\nAny mismatches at all? ->", any_mismatch)
```

```python
import pandas as pd
import numpy as np

# df_export: this is the 106-row table you want to see
df_export = moca_wide_filtered.copy()

both_mask = merged['_merge'] == 'both'

# Dict: for each wide-column, a boolean Series (index aligned with df_export)
new_flags = {col_wide: pd.Series(False, index=df_export.index) 
             for _, (_, col_wide) in pairs.items()}

for name, (col_moca, col_wide) in pairs.items():
    if col_moca not in merged.columns or col_wide not in merged.columns:
        continue

    left  = pd.to_numeric(merged[col_moca], errors='coerce')
    right = pd.to_numeric(merged[col_wide], errors='coerce')

    # New value = old Moca NaN, wide has a number
    mask_new = both_mask & left.isna() & right.notna()

    # Which participants have a new value for this variable?
    ids_with_new = merged.loc[mask_new, 'id_participant']

    # Mark those rows as True in the flags, aligned on df_export
    new_flags[col_wide][df_export['id_participant'].isin(ids_with_new)] = True

```

```python
output_path = "C:/Users/okkam/Desktop/labo/article 2/Moca/moca_wide_with_new_values_bold.xlsx"

with pd.ExcelWriter(output_path, engine="xlsxwriter") as writer:
    # Write the plain table (no flags)
    df_export.to_excel(writer, index=False, sheet_name="Moca")

    workbook  = writer.book
    worksheet = writer.sheets["Moca"]

    bold_fmt = workbook.add_format({'bold': True})

    # For each column we track, overwrite ONLY the "new" cells in bold
    for col_wide, flags in new_flags.items():
        if col_wide not in df_export.columns:
            continue

        col_idx = df_export.columns.get_loc(col_wide)

        for row_idx, is_new in flags.items():
            if is_new:
                value = df_export.iloc[row_idx, col_idx]
                # +1 because row 0 in Excel is the header
                worksheet.write(row_idx + 1, col_idx, value, bold_fmt)

print("Saved:", output_path)
```
