```python
import pandas as pd

# Input files
left_file = r"C:\Users\okkam\Desktop\labo\article 2\Longitudinal_Multimodal_Data_CIMAQ\2026-04-04_glu_moca_struc.xlsx"
right_file = r"C:\Users\okkam\Desktop\labo\article 2\Longitudinal_Multimodal_Data_CIMAQ\Glu\databases_previous\import_data_arsenii_20250609_hippR_article1.xlsx"

# Output file (saved next to the left table)
output_file = r"C:\Users\okkam\Desktop\labo\article 2\Longitudinal_Multimodal_Data_CIMAQ\2026-04-04_glu_moca_struc_left_join_article1.xlsx"

# Read Excel files
left_df = pd.read_excel(left_file)
right_df = pd.read_excel(right_file)

# Ensure key columns are strings and clean whitespace
left_df["PSCID"] = left_df["PSCID"].astype(str).str.strip()
right_df["PSCID_CandID"] = right_df["PSCID_CandID"].astype(str).str.strip()

# Extract PSCID from PSCID_CandID (text before first underscore)
right_df["PSCID"] = right_df["PSCID_CandID"].str.split("_", n=1).str[0].str.strip()

# Left join: keep all rows from left table
merged_df = left_df.merge(right_df, on="PSCID", how="left")

# Save result
merged_df.to_excel(output_file, index=False)

print(f"Left join complete. File saved to:\n{output_file}")
print(f"Rows in left table: {len(left_df)}")
print(f"Rows in merged table: {len(merged_df)}")
print("Columns in merged table:")
print(list(merged_df.columns))
```
